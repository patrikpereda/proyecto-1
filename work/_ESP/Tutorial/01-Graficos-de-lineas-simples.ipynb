# Simple Line Plots

Quizás el más simple de todos los gráficos es la visualización de una sola función $y = f(x)$.
Aquí veremos por primera vez cómo crear un gráfico simple de este tipo.
Como en todas las secciones siguientes, comenzaremos configurando el cuaderno para realizar gráficos e importando los paquetes que usaremos:

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
import numpy as np
````

Para todos los gráficos de Matplotlib, comenzamos creando una figura y un eje.
En su forma más simple, una figura y un eje se pueden crear de la siguiente manera:

````python
fig = plt.figure()
ax = plt.axes()
````

En Matplotlib, la *figura* (una instancia de la clase ``plt.Figure``) puede considerarse como un único contenedor que contiene todos los objetos que representan ejes, gráficos, texto y etiquetas.
Los *ejes* (una instancia de la clase ``plt.Axes``) son lo que vemos arriba: un cuadro delimitador con marcas y etiquetas, que eventualmente contendrá los elementos del gráfico que componen nuestra visualización.
A lo largo de este libro, utilizaremos comúnmente el nombre de variable ``fig`` para referirnos a una instancia de figura, y ``ax`` para referirnos a una instancia de ejes o un grupo de instancias de ejes.

Una vez que hemos creado los ejes, podemos utilizar la función ``ax.plot`` para representar gráficamente algunos datos. Comencemos con una sinusoide simple:

````python
fig = plt.figure() #Figura
ax = plt.axes()    #Ejes

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
````

Alternativamente, podemos utilizar la interfaz de pylab y dejar que la figura y los ejes se creen en segundo plano:

````python
plt.plot(x, np.sin(x));
````

Si queremos crear una sola figura con varias líneas, simplemente podemos llamar a la función ``plot`` varias veces:

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
````

¡Eso es todo lo que hay que saber para trazar funciones simples en Matplotlib!
Ahora analizaremos más en detalle cómo controlar la apariencia de los ejes y las líneas.

## Adjusting the Plot: Line Colors and Styles

El primer ajuste que puede que desee realizar en un gráfico es controlar los colores y estilos de las líneas.
La función ``plt.plot()`` acepta argumentos adicionales que pueden utilizarse para especificarlos.
Para ajustar el color, puede utilizar la palabra clave ``color``, que acepta un argumento de cadena que representa prácticamente cualquier color imaginable.
El color se puede especificar de diversas formas:

````python
plt.plot(x, np.sin(x - 0), color='blue')        # especifica el color por nombre
plt.plot(x, np.sin(x - 1), color='g')           # código de color corto (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#97EF09')     # Código hexadecimal (RRGGBB de 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Tupla RGB, valores de 0 a 1.
plt.plot(x, np.sin(x - 5), color='chartreuse'); # todos los nombres de colores HTML admitidos
````

````python
for i in range(6):
  plt.plot(x, np.sin(x - i))
````

Si no se especifica ningún color, Matplotlib recorrerá automáticamente un conjunto de colores predeterminados para varias líneas.

De manera similar, el estilo de línea se puede ajustar utilizando la palabra clave ``linestyle``:

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Para abreviar, puedes utilizar los siguientes códigos:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

Si desea ser extremadamente conciso, estos códigos ``linestyle`` y ``color`` se pueden combinar en un solo argumento que no sea una palabra clave para la función ``plt.plot()``:

````python
plt.plot(x, x + 0, '-g') # verde sólido
plt.plot(x, x + 1, '--c') # cian discontinuo
plt.plot(x, x + 2, '-.k') # negro discontinuo
plt.plot(x, x + 3, ':r'); # rojo punteado
````

Estos códigos de color de un solo carácter reflejan las abreviaturas estándar de los sistemas de color RGB (rojo/verde/azul) y CMYK (cian/magenta/amarillo/negro), que se utilizan habitualmente para gráficos en color digitales.

Hay muchos otros argumentos de palabras clave que se pueden utilizar para ajustar la apariencia del gráfico; para obtener más detalles, sugiero ver el docstring de la función ``plt.plot()`` utilizando las herramientas de ayuda de IPython

## Adjusting the Plot: Axes Limits

Matplotlib hace un buen trabajo al elegir los límites de los ejes predeterminados para su gráfico, pero a veces es bueno tener un control más preciso.
La forma más básica de ajustar los límites de los ejes es usar los métodos ``plt.xlim()`` y ``plt.ylim()``:

````python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
````

Si por alguna razón desea que alguno de los ejes se muestre al revés, puede simplemente invertir el orden de los argumentos:

````python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
````

Un método útil relacionado es ``plt.axis()`` (nótese aquí la posible confusión entre *axes* con una *e* y *axis* con una *i*).
El método ``plt.axis()`` le permite establecer los límites ``x`` e ``y`` con una única llamada, pasando una lista que especifica ``[xmin, xmax, ymin, ymax]``:

````python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
````

El método ``plt.axis()`` va incluso más allá de esto, permitiéndole hacer cosas como ajustar automáticamente los límites alrededor del gráfico actual:

````python
plt.plot(x, np.sin(x))
plt.axis('tight');
````

Permite especificaciones de nivel aún más alto, como garantizar una relación de aspecto igual para que en la pantalla, una unidad en ``x`` sea igual a una unidad en ``y``:

````python
plt.plot(x, np.sin(x))
plt.axis('equal');
````

Para obtener más información sobre los límites de los ejes y otras capacidades del método ``plt.axis``, consulte la cadena de documentación ``plt.axis``.

## Labeling Plots

Como última parte de esta sección, analizaremos brevemente el etiquetado de los gráficos: títulos, etiquetas de ejes y leyendas simples.

Los títulos y las etiquetas de ejes son las etiquetas más simples de este tipo; existen métodos que se pueden utilizar para configurarlas rápidamente:

````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve") # Titulo
plt.xlabel("x")           # Etiqueta en eje X  
plt.ylabel("sin(x)");     # Etiqueta en eje Y  
````

La posición, el tamaño y el estilo de estas etiquetas se pueden ajustar mediante argumentos opcionales para la función.
Para obtener más información, consulte la documentación de Matplotlib y las docstring de cada una de estas funciones.

Cuando se muestran varias líneas dentro de un solo eje, puede resultar útil crear una leyenda gráfica que etiquete cada tipo de línea.
Nuevamente, Matplotlib tiene una forma incorporada de crear rápidamente una leyenda de este tipo.
Se hace a través del método (lo adivinaste) ``plt.legend()``.
Aunque hay varias formas válidas de usar esto, me parece más fácil especificar la etiqueta de cada línea usando la palabra clave ``label`` de la función de gráfico:

````python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
````

Como puede ver, la función ``plt.legend()`` realiza un seguimiento del estilo y el color de la línea y los combina con la etiqueta correcta.
Puede encontrar más información sobre cómo especificar y formatear leyendas de gráficos en ``plt.legend``

## Aside: Matplotlib Gotchas

Si bien la mayoría de las funciones ``plt`` se traducen directamente a métodos ``ax`` (como ``plt.plot()`` → ``ax.plot()``, ``plt.legend()`` → ``ax.legend()``, etc.), este no es el caso de todos los comandos.
En particular, las funciones para establecer límites, etiquetas y títulos se modifican ligeramente.
Para realizar la transición entre funciones de estilo MATLAB y métodos orientados a objetos, realice los siguientes cambios:

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

En la interfaz orientada a objetos para trazar gráficos, en lugar de llamar a estas funciones individualmente, a menudo es más conveniente utilizar el método ``ax.set()`` para configurar todas estas propiedades a la vez:

````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
````
